```python
# 추가된 경로에서 모듈 임포트
from gcpVision import get_vision_ocr, get_image_search
from imDetect import cropImg
```

```python
# 이미지를 불러옵니다.(실제 시스템에서는 동일 방법으로 불러와졌다고 가정함)
file_name = './SampleImages/BlogWithImages.jpg'

# bytes 형태로 불러와 진 것을 가정합니다.
with io.open(file_name, 'rb') as image_file:
    content = image_file.read()
```

```python
# 글자 인식을 수행하여 반환합니다.
ocr = get_vision_ocr(content)
print(ocr)
```

```python
# 이미지에서 사진 부분을 추론하여 잘라냅니다.
img = cropImg(content)
print(len(img)) # 바이트 원소의 배열이 반환되어 길이로 대체합니다.
```

```python
# (잘린 이미지 시각화) 바이트 이미지를 디코딩하여 시각화
dec_img = np.frombuffer(img[0], dtype=np.uint8)
dec_img = cv2.imdecode(dec_img, flags=1)

plt.imshow(dec_img)
plt.show()
```

```python
# url 주소 검색
urls = get_image_search(img)
urls # 이미지 개수의 인덱스를 가진 리스트가 반환됩니다.
```
